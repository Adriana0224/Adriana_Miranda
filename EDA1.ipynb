# **Filtrado de datos**

```python
Total_Online = df2[df2["Origen"]=="Online"]
VO= sum(Total_Online["Total"])
print("El total en Online es---->",VO)

Total_Fisica = df2[df2["Origen"]== "Tienda Física"]
VF =sum(Total_Fisica["Total"])
print("El total en Tienda Fisica es---->",VF)
```

```python
x=["Online", "Tienda Física"]
y=[VO,VF]

fig, ax = plt.subplots()
ax.bar(x,y,linewidth=20)

ax.set_title("Totales ventas por tipo de tienda", fontsize= 14)
ax.set_xlabel("Tipo de tienda", fontsize = 12)
ax.set_ylabel("Total ventas", fontsize =12)
```

```python
TotalesOrigen =df2[["Origen","Total"]]
TotalesOrigen
TotalesOrigen2 = TotalesOrigen.groupby("Origen").sum()
TotalesOrigen2

fig, ax =plt.subplots()
valoresm = TotalesOrigen2.Total
titulosm = TotalesOrigen2.index
coloresm = ["#4DAF4A", "#377EB8"]

ax.bar(titulosm,valoresm,color = coloresm)

ax.set_title("Totales ventas por tipo de tienda", fontsize= 14)
ax.set_xlabel("Tipo de tienda", fontsize = 12)
ax.set_ylabel("Total ventas", fontsize =12)
```

```python
Cantidad_ventas = df2[["Cliente","Origen"]]
Cantidad_ventas
Cantidad_ventas2 = Cantidad_ventas.groupby("Origen").count()
Cantidad_ventas2

fig,ax =plt.subplots()
valoresa= Cantidad_ventas2.Cliente
titulosa= Cantidad_ventas2.index
coloresa=["#4DAF4A", "#377EB8"]

ax.bar(titulosa, valoresa, color = coloresa)
```

```python
Cantida_Online =df2[df2["Origen"]=="Online"]
Cant2=Cantida_Online["Origen"].count()
print("La cantidad de ventas Online es----->",Cant2)

Cantidad_Fisica = df2[df2["Origen"]== "Tienda Física"]
Cant3 = Cantidad_Fisica["Origen"].count()
print("La cantidad de ventas en Tienda Física es----->",Cant3)
```

```python
fig,ax=plt.subplots()
Titulos=["Online","Tienda Física"]
Dato=[Cant2,Cant3]


ax.pie(Dato, labels=Titulos, explode= [0.05,0], autopct ="%1.0f%%")
ax.set_title("Porcentaje de ventas en cada tienda", fontsize=14)
```

```python
df2[["Producto","Total"]]
prod = df2[["Producto","Total"]]
prod1 = prod.groupby("Producto").sum()

fig,ax=plt.subplots()
Titulos=prod1.index
Dato=prod1.Total

ax.pie(Dato, labels=Titulos, autopct="%1.1f%%")
ax.set_title("Porcentaje de ventas por productos", fontsize=14)
```
