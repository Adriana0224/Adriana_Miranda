```python
# Crear datos con NumPy
x = np.linspace(0, 10, 1000)
y = np.sin(x)

# Usar Seaborn para un estilo bonito
fig,ax = plt.subplots()

# Graficar con Matplotlib
plt.plot(x, y)
plt.title("Gráfico combinado usando numpy, seaborn y matplotlib")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

```python
# Crear datos
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)


# Crear figura con 2 filas y 2 columnas de gráficos
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Primer gráfico: seno (fila 0, columna 0)
axes[0].plot(x, y1, color='blue')
axes[0].set_title("Seno")
axes[0].set_xlabel("x")
axes[0].set_ylabel("sin(x)")

# Segundo gráfico: coseno (fila 0, columna 1)
axes[1].plot(x, y2, color='green')
axes[1].set_title("Coseno")
axes[1].set_xlabel("x")
axes[1].set_ylabel("cos(x)")

# Ajustar el diseño
plt.tight_layout()
plt.show()
```

```python
# Crear datos
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Crear figura con 2 filas y 2 columnas de gráficos
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Primer gráfico: seno (fila 0, columna 0)
axes[0].plot(x, y1, color='blue')
axes[0].set_title("Seno")
axes[0].set_xlabel("x")
axes[0].set_ylabel("sin(x)")

# Segundo gráfico: coseno (fila 0, columna 1)
axes[1].plot(x, y2, color='green')
axes[1].set_title("Coseno")
axes[1].set_xlabel("x")
axes[1].set_ylabel("cos(x)")

# Tercer gráfico: tangente (fila 1, columna 0)
axes[2].plot(x, y3, color='orange')
axes[2].set_title("Tangente")
axes[2].set_xlabel("x")
axes[2].set_ylabel("tan(x)")

# Ajustar el diseño
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Crear datos
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Crear figura con 2 filas y 2 columnas de gráficos
fig, axes = plt.subplots(2, 2, figsize=(12, 6))

# Primer gráfico: seno (fila 0, columna 0)
axes[0, 0].plot(x, y1, color='blue')
axes[0, 0].set_title("Seno")
axes[0, 0].set_xlabel("x")
axes[0, 0].set_ylabel("sin(x)")

# Segundo gráfico: coseno (fila 0, columna 1)
axes[0, 1].plot(x, y2, color='green')
axes[0, 1].set_title("Coseno")
axes[0, 1].set_xlabel("x")
axes[0, 1].set_ylabel("cos(x)")

# Tercer gráfico: tangente (fila 1, columna 0)
axes[1, 0].plot(x, y3, color='orange')
axes[1, 0].set_title("Tangente")
axes[1, 0].set_xlabel("x")
axes[1, 0].set_ylabel("tan(x)")

# Cuarto gráfico vacío (fila 1, columna 1) opcionalmente lo dejas sin usar o le pones algo
axes[1, 1].axis('off')  # Ocultamos ese panel

# Ajustar el diseño
plt.tight_layout()
plt.show()
```
